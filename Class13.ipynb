```python
from agents import enable_verbose_stdout_logging, function_tool

enable_verbose_stdout_logging()

@function_tool
def get_current_weather(city: str) -> str:
  """ Return Weather for the city

  Args:
    city: City Name

  Returns:
   Weather: str
  """
  return f"{city} the weather is Sunny"

@function_tool
def get_current_news(industry: str) -> str:
  """ Return current news for a particular industry

  Args:
    industry: The industry to look in current News

  Returns:
   news
  """
  return f"{industry} have new concept Agent Native Cloud "

agent: Agent = Agent(name="Assistant",
                     instructions="You are a helpful assistant",
                     model="gemini-1.5-flash",
                     tools=[get_current_weather, get_current_news]
                    )
result = Runner.run_sync(agent, "What is the weather of Karachi and share latest news in Tech")

print(result.final_output)
```

```python
from agents import enable_verbose_stdout_logging, function_tool, Agent, Runner

enable_verbose_stdout_logging()

@function_tool
def get_current_weather(city: str) -> str:
  """ Return Weather for the city

  Args:
    city: City Name

  Returns:
   Weather: str
  """
  return f"{city} the weather is Sunny"

@function_tool
def get_current_news(industry: str) -> str:
  """ Return current news for a particular industry

  Args:
    industry: The industry to look in current News

  Returns:
   news
  """
  return f"{industry} have new concept Agent Native Cloud "

panaversity_agent: Agent = Agent(name="Panaversity Assistant",
                     instructions="You are a helpful panavesity assistant. You answer all about Panaversity",
                     model="gemini-1.5-flash",
                     handoff_description="Panaversity Expert"
                     )

agentic_ai_agent: Agent = Agent(name="Agentic AI Assistant",
                     instructions="You are a helpful agentic ai expert. You answer all about AI agents",
                     model="gemini-1.5-flash",
                     handoff_description="Agentic AI Expert"
                    )


triage_agent: Agent = Agent(
    name="Triage Agent",
    instructions="You are a general chat assistant.You observe the communication with user and let relevant agants answer back to the user quiries",
    model="gemini-1.5-flash",
    # tools=[get_current_weather, get_current_news],
    handoffs=[panaversity_agent, agentic_ai_agent]
)

result = Runner.run_sync(triage_agent, "Who is the founder of PIAIC?")

print(result.final_output)
```
